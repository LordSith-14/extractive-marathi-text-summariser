```python
def readStopWords():
    try:
        with io.open('stopwords.txt',encoding='utf-8') as textFile:
            for line in textFile:
                words = line.lower().strip()
                stopwords.append(words)
            textFile.close()
    except IOError:
        print("error")
    return stopwords
        
```

```python
def stopword_removal(text):
    
    #Tokenize the text
    words = word_tokenize(text)
    
    # Removal of the Stopwords 
    # By using the stopwords in the marathiST.txt file    
    filtered_words = [word for word in words if word not in readStopWords()]
    # In future more stopwords can be added to the file according to the Requirnment
    
    # Join the filtered text back to obtain an string
    text_without_stp_words = ' '.join(filtered_words)
    
    return text_without_stp_words
```

```python
def remove_marathi_punctuation(text):
    marathi_punctuation = '।‘’!"#$%&\'()*+,-/:;<=>?@[\\]^_`{|}~'
    # We may can extend this list based on your specific requirements
    
    # Regular expression pattern to match Marathi punctuation
    pattern = '[' + re.escape(marathi_punctuation) + ']'
    
    # Remove Marathi punctuation using the pattern
    text_without_punctuation = re.sub(pattern, '', text)
    return text_without_punctuation
```

```python
def readStemWords():
    global stemWords
    try:
        with io.open("word_list_marathi.txt", encoding='utf-8') as textFile:
            for line in textFile:
                if len(line):
                    wordEndIndex = line.find(">")
                    word = line[1:wordEndIndex]
                    line = line[wordEndIndex+1:]
                    baseEndIndex = line.find("]")
                    base = line[1:baseEndIndex].strip()
                    line = line[baseEndIndex+1:]
                    stem = None
                    if len(line)>0:
                        stemEndIndex = base.find("-")
                        if stemEndIndex>0:
                            stem = base[:stemEndIndex]
                    line = line[line.find("{")+1:line.find("}")].strip()
                    related=[]
                    if len(line)>0:
                        r1 = line.split(",")
                        for s in r1:
                            related.append(s[:s.find("|")])
                    if stem == None and len(related)>0:
                        stem = related[0]
                    if stem !=None:
                        stemWords[word] = {}
                        stemWords[word]["stem"] = stem
                        stemWords[word]["related"] = related                         
    except IOError:
        print("Error: File cannot be Opened")
    return stemWords    
```

```python
def stemmerMarathi(text):
    return [removeNoGender(word) for word in text]
```

```python
text = "वर्तमान गटामध्ये आपलं स्वप्न पूर्ण होणार आहे. आपलं असा परिपूर्ण समुदाय तयार करण्यासाठी, आपण सर्वांचं सहयोग कडेच कायम करावं लागेल."
def restoring_stem(text): 
    word_token = word_tokenize(text)
    stem_words = stemmerMarathi(word_token)
    cln_text = ' '.join(stem_words)
    return cln_text
```

```python
def tokenize(filepath):
    global sentences,sentence_dictionary,sentence_processing 
    with io.open(filepath,encoding='utf-8') as inputFile:
        text = inputFile.read()
        inputFile.close()
    text_without_puntuation = remove_marathi_punctuation(text)
    text_without_stopwords = stopword_removal(text_without_puntuation)
    clean_text = restoring_stem(text_without_stopwords)
    
    sentences = sent_tokenize(clean_text)
    sentence_processing = copy.deepcopy(sentences)
    counter = 0
    for sentence in sentence_processing:
#         print(sentence)
        sentence = re.sub(',|\.|-|\(|\)', ' ', sentence)
        tokens = sentence.strip().split()
#         print(tokens)
        sentence_dictionary[counter] = tokens
        counter+=1
```

```python
def cleanText(filename):
    '''
        Tokenize, Remove stopwords and reduce the words to their stem
    :param filename: path of file to be preprocessed
    '''
    global sentence_dictionary, sentences
    readStopWords()
    tokenize(filename)
    size = 0
    for i in range(0, len(sentence_dictionary)):
        size += len(sentence_dictionary[i])
    sentence_dictionary = {key: value for key,
                           value in sentence_dictionary.items() if len(value) > 0}
    return sentence_dictionary, sentences, size
```
